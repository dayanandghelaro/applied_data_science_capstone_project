### 2.2 - Import Necessary Packages/Libraries

```python
import pandas as pd # library for data analsysis
import numpy as np # library to handle data in a vectorized manner
import json # library to handle JSON files
import requests # library to handle requests
from pandas.io.json import json_normalize # tranform JSON file into a pandas dataframe
# packages to scrape the data from webpage
from bs4 import BeautifulSoup
import lxml 
import html5lib

from geopy.geocoders import Nominatim # module to convert an address into latitude and longitude values

import folium # plotting library

# Matplotlib and associated plotting modules
import matplotlib.cm as cm
import matplotlib.colors as colors

# import k-means from clustering stage
from sklearn.cluster import KMeans
```

### 2.3 - Get Union Names

```python
source = requests.get('https://en.wikipedia.org/wiki/List_of_Union_Councils_of_Karachi').text
soup = BeautifulSoup(source,'lxml')

data_unions = soup.find('div', class_="mw-parser-output")
union_names = []

for e in data_unions.find_all('h3'):
    union_names.append(e.text.replace('[edit]',""))

print(union_names)

```

### 2.4 - Get Town Names

```python
town_names = []
data_towns = [[],[]]
indx= -1
for data_towns_tables in data_unions.find_all('table'):
    if indx == -1:
        indx = 0
        continue
    else:
        table_body = data_towns_tables.tbody
        for record in table_body.find_all('td'):      
            if record.find('a') is not None:
                record_ = record.find('a').text
                town_names.append(record_)
                data_towns[0].append(record_)
                data_towns[1].append(union_names[indx-1])        
        indx+=1
#        print(indx)
#print(len(union_names))
print(town_names)
```

### 2.5 - Creating Dictionary of Data

```python
data_dict={}
data_dict['town_name'] = data_towns[0]
data_dict['union_name'] = data_towns[1]
```

### 2.6 - Creating Data Frame (for easy manipulation)

```python
data_frame = pd.DataFrame(data_dict)
data_frame.head(10)
```

### 2.8 - Getting Latitudes and Longitudes of Unions

```python
latitudes_unions = {}
longitudes_unions = {}
for i in range(len(union_names)):
    address = union_names[i]+", Karachi"
    geolocator = Nominatim(user_agent="foursquare_agent")
    location = geolocator.geocode(address)
    if location is not None:
        latitude = location.latitude
        longitude = location.longitude
    else:
        latitude = "None"
        longitude = "None"
    latitudes_unions[union_names[i]] = latitude
    longitudes_unions[union_names[i]] = longitude
#    latitudes_unions.append(latitude)
#    longitudes_unions.append(longitude)
    print(address+"\t"+str(latitude)+"\t"+str(longitude))
```

### 2.9 - Getting Latitudes and Longitudes of Towns

```python
latitudes = []
longitudes= []
for i in range(len(town_names)):
    address = data_towns[0][i]+", "+ data_towns[1][i]+", Karachi"
    geolocator = Nominatim(user_agent="foursquare_agent")
    location = geolocator.geocode(address)
    if location is not None:
        latitude = location.latitude
        longitude = location.longitude
    else:
        latitude = latitudes_unions[data_towns[1][i]]
        longitude = longitudes_unions[data_towns[1][i]]
    latitudes.append(latitude)
    longitudes.append(longitude)
    print(address+"\t"+str(latitude)+"\t"+str(longitude))
```

### 2.10 - Creating Updated Data Frame

```python
data_dict={}
data_dict['town_name'] = data_towns[0]
data_dict['union_name'] = data_towns[1]
data_dict['latitude'] = latitudes
data_dict['longitude'] = longitudes

data_frame = pd.DataFrame(data_dict)
data_frame.head(10)
```

## 3. Methodology

### 3.1 - Map of Karachi and Neighborhoods

```python
# create map of Karachi using latitude and longitude values
map_karachi = folium.Map(location=[latitude, longitude], zoom_start=12)

# add markers to map
for lat, lng, borough, neighborhood in zip(data_frame['latitude'], data_frame['longitude'], data_frame['union_name'], data_frame['town_name']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_karachi) 
    folium.Marker([lat, lng],
                 popup=data_frame['town_name']).add_to(map_karachi)
    
map_karachi
```

### 3.3 - Clustering

```python
data_frame.head()
```

```python
# set number of clusters
kclusters = 5

clustering = data_frame[['latitude','longitude']]

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

```python
data_frame.insert(0, 'cluster_label', kmeans.labels_)
data_frame.head()
```

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(data_frame['latitude'], data_frame['longitude'], data_frame['town_name'], data_frame['cluster_label']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
